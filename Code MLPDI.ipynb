<h2>Question 8</h2>

<span style="color:#1E6AFF;font-family:trebuchet ms;">
On commence par simuler le processus $X_n$ un nombre n de fois, que l'utilisateur doit renseigner au préalable.

```python
n=input('Entrez la valeur de n = ')
n=int(n)

G=0
D=0
X=[]
X.append(0)

for i in range(1,n+1):
    U=rd.uniform(0,1)
    if U>(i+2+X[-1])/(2*(i+2)):
        D=D+1
    else:
        G=G-1
    X.append(G+D)

figure(figsize=(20,10))
plot(X,linewidth=3)
title('Simulation du processus X',fontsize=30, color="blue")
savefig("Simulation processus X proba 0.5.png", format="png")
show()    
```

<h2>Question 9</h2>

<span style="color:#1E6AFF;font-family:trebuchet ms;">
On reprend le tout premier code en changeant la probabilité d'aller à droite ou à gauche (qui était de 1/2 avant) pour permettre à l'utilisateur de choisir la probabilité p d'aller vers la droite (et donc (1-p) pour la gauche).

```python
n=input('Entrez la valeur de n = ')
n=int(n)
p=input('Entrez la valeur de p = ')
p=float(p)

if p==0.5 or p==0 or p==1:
    print("Ne pas prendre p=1/2 ou p=1 ou p=0.")

G=0
D=0
X=[]
X.append(0)

for i in range(1,n+1):
    U=rd.uniform(0,1)
    k1=1-((1-p)/p)**((X[-1]-n)/2-1)
    k2=((1-p)/p)**((X[-1]+n)/2+1)-((1-p)/p)**((X[-1]-n)/2-1)
    k=k1/k2
    if U<k:
        D=D+1
    else:
        G=G-1
    X.append(G+D)

figure(figsize=(20,10))
plot(X,linewidth=3)
title('Simulation du processus X',fontsize=30, color="blue")
savefig("Simulation processus X proba 0.51.pdf", format="pdf")
show()    
```

Voici le code qui permet de simuler une seule marche aléatoire simple sur $\mathbb{Z}^2$. Nous avons rajouté une croix rouge au point de départ situé en {0;0}.

```python
X=[]
Y=[]

X.append(0)
Y.append(0)

n=100000

for i in range(1,n+1):
    U=rd.uniform(0,1)
    if U>=1/2:
        if U>=3/4:
            X.append(X[-1]+1)
        else:
            X.append(X[-1]-1)
        Y.append(Y[-1])
    if U<1/2:
        if U>=1/4:
            Y.append(Y[-1]+1)
        else:
            Y.append(Y[-1]-1)
        X.append(X[-1])
        
figure(figsize=(20,10))
plot(X,Y,linewidth=3)
plot([-1,1],[-1,1],linewidth=3,color="red")
plot([-1,1],[1,-1],linewidth=3,color="red")
title('Simulation marche aléatoire dans Z²',fontsize=30, color="blue")
savefig("Simulation marche aléatoire dans Z2.png", format="png")
show() 
```

<span style="color:#1E6AFF;font-family:trebuchet ms;">

Ce code reprend le précédent mais cette fois-ci, il permet de réaliser N fois une marche aléatoire avec une probabilité p de se déplacer horizontalement et (1-p) verticalement.

```python
def marcheZ2(p):

    clf()

    

    n=1000
    N=1000
    
    if p>1/2:
        figure(figsize=(30,20))
    elif p==1/2:
        figure(figsize=(20,20))
    else:
        figure(figsize=(20,30))

    for l in range (N):
        
        X=[]
        Y=[]

        X.append(0)
        Y.append(0)

        for i in range(1,n+1):
            U=rd.uniform(0,1)
            if U>=p:
                if U>=p+(1-p)/2:
                    Y.append(Y[-1]+1)
                else:
                    Y.append(Y[-1]-1)
                X.append(X[-1])
            if U<p:
                if U>=p/2:
                    X.append(X[-1]+1)
                else:
                    X.append(X[-1]-1)
                Y.append(Y[-1])
    
        
        plot(X,Y,linewidth=3)
        
    title('Simulation marche aléatoire dans Z²',fontsize=30, color="blue")
    savefig("Simulation N marches aléatoires dans Z2 avec p="+str(p)+".png", format="png")
    show()
    clf()
```
